# 4

```python
black_duck = skimage.color.rgb2gray(duckling)
print(black_duck.shape)
io.imshow(black_duck)
plt.title('black_duck')
plt.show()

cool_avg_duck = np.mean(duckling[:,:,:3], axis=2)
print(cool_avg_duck.shape)
#cool_avg_duck = cool_avg_duck.astype('uint8')
io.imshow(cool_avg_duck)
plt.title('cool_avg_duck')
plt.show()

concat = np.concatenate((black_duck, cool_avg_duck), axis=1)
io.imshow(concat, cmap='Greys_r')
plt.title('rgb2gray vs. np.mean')
plt.show()
```
